2. **One-hot encoding**: Terapkan pada fitur dengan jumlah nilai unik yang relatif kecil dan tidak memiliki urutan logis, seperti job, marital, contact, poutcome, education, month, day_of_week, dan age_bin. One-hot encoding mengubah setiap kategori menjadi kolom biner terpisah (0 atau 1), sehingga model dapat menangani fitur kategorikal tanpa memperhitungkan urutan antar kategori.

3. **Label encoding**: Gunakan untuk fitur yang memiliki nilai unik dengan urutan atau relasi ordinal, seperti contact_quarter, emp.var.rate_bin, cons.price.idx_bin, cons.conf.idx_bin, euribor3m_bin, nr.employed_bin, serta default, housing, dan loan. Label encoding memberikan representasi numerik yang mempertahankan urutan atau relasi ordinal di antara kategori.

# **<span style='color:#E1B12D'><i> MODELLING </i></span>**

**Metrik Evaluasi dalam Konteks Bisnis Deposit Bank**

1. **Precision:** Proporsi prediksi positif yang benar.
2. **Recall:** Proporsi aktual positif yang berhasil diprediksi.
3. **F1-Score:** Harmonic mean dari Precision dan Recall.
4. **ROC-AUC:** Mengukur kemampuan model membedakan kelas positif dan negatif.
5. **Accuracy:** Proporsi total prediksi yang benar.

**Fokus pada Recall dan ROC-AUC**

* **Recall:** Sangat penting untuk meminimalkan kehilangan peluang bisnis. Dengan Recall tinggi, bank dapat mengidentifikasi sebagian besar pelanggan potensial.
* **ROC-AUC:** Ideal untuk data tidak seimbang dan memberikan gambaran menyeluruh tentang kinerja model.

**Penjelasan Lebih Lanjut**

* **Recall:** Jika Recall tinggi, berarti model berhasil mengidentifikasi sebagian besar pelanggan yang benar-benar akan membuka deposit. Ini penting karena kita tidak ingin kehilangan pelanggan potensial.
* **ROC-AUC:** ROC-AUC memberikan gambaran yang lebih komprehensif dibandingkan Accuracy, terutama ketika kelas data tidak seimbang (misalnya, jumlah pelanggan yang membuka deposit jauh lebih sedikit). ROC-AUC mengukur kemampuan model dalam membedakan antara kelas positif (akan membuka deposit) dan negatif (tidak akan membuka deposit) untuk berbagai kemungkinan threshold.



```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.model_selection import cross_validate
import pandas as pd
import matplotlib.pyplot as plt

def eval_classification(model, X_train_resampled, y_train_resampled, X_test, y_test):
    # Prediksi pada set test dan train
    y_pred = model.predict(X_test)
    y_pred_train = model.predict(X_train_resampled)

    # Probabilitas prediksi pada set test dan train
    y_pred_proba = model.predict_proba(X_test)
    y_pred_proba_train = model.predict_proba(X_train_resampled)

    # Menghitung dan mencetak metrik evaluasi untuk set test
    print("Accuracy (Test Set): %.2f" % accuracy_score(y_test, y_pred))
    print("Accuracy (Train Set): %.2f" % accuracy_score(y_train_resampled, y_pred_train))

    print("Precision (Test Set): %.2f" % precision_score(y_test, y_pred))
    print("Precision (Train Set): %.2f" % precision_score(y_train_resampled, y_pred_train))

    print("Recall (Test Set): %.2f" % recall_score(y_test, y_pred))
    print("Recall (Train Set): %.2f" % recall_score(y_train_resampled, y_pred_train))

    print("F1-Score (Test Set): %.2f" % f1_score(y_test, y_pred))
    print("F1-Score (Train Set): %.2f" % f1_score(y_train_resampled, y_pred_train))

    # Menghitung dan mencetak ROC-AUC untuk set test
    print("ROC-AUC (Test Set - Proba): %.2f" % roc_auc_score(y_test, y_pred_proba[:, 1]))
    print("ROC-AUC (Train Set - Proba): %.2f" % roc_auc_score(y_train_resampled, y_pred_proba_train[:, 1]))

    # Cross-validation untuk Precision, Recall, F1-Score, dan ROC-AUC
    cv_results = cross_validate(
        model,
        X_train_resampled,
        y_train_resampled,
        cv=5,
        scoring={'precision': 'precision', 'recall': 'recall', 'f1': 'f1', 'roc_auc': 'roc_auc'},
        return_train_score=True
    )

    print('Precision (Cross-Val Train): %.2f' % cv_results['train_precision'].mean())
    print('Precision (Cross-Val Test): %.2f' % cv_results['test_precision'].mean())
    print('Recall (Cross-Val Train): %.2f' % cv_results['train_recall'].mean())
    print('Recall (Cross-Val Test): %.2f' % cv_results['test_recall'].mean())
    print('F1-Score (Cross-Val Train): %.2f' % cv_results['train_f1'].mean())
    print('F1-Score (Cross-Val Test): %.2f' % cv_results['test_f1'].mean())
    print('ROC-AUC (Cross-Val Train): %.2f' % cv_results['train_roc_auc'].mean())
    print('ROC-AUC (Cross-Val Test): %.2f' % cv_results['test_roc_auc'].mean())

def show_feature_importance(model, X):
    feat_importances = pd.Series(model.feature_importances_, index=X.columns)
    ax = feat_importances.nlargest(60).plot(kind='barh', figsize=(10, 8))
    ax.invert_yaxis()

    plt.xlabel('score')
    plt.ylabel('feature')
    plt.title('Feature Importance Score')
    plt.show()

def show_best_hyperparameter(model):
    if hasattr(model, 'best_estimator_'):
        print(model.best_estimator_.get_params())
    else:
        print("Model tidak memiliki atribut 'best_estimator_'")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.metrics import roc_auc_score

def draw_learning_curve(param_values, model_name, model_class, param_name, param_range):
    train_scores = []
    test_scores = []

    for param_value in param_values:
        if model_name == 'SVM':
            model = model_class(C=param_value, probability=True)  # Gunakan 'probability=True' untuk mendapatkan probabilitas
        else:
            model = model_class(**{param_name: param_value})
        model.fit(X_train_resampled, y_train_resampled)

        # Evaluasi pada data pelatihan
        y_pred_train_proba = model.predict_proba(X_train_resampled)[:, 1] if model_name != 'SVM' else model.decision_function(X_train_resampled)
        train_auc = roc_auc_score(y_train_resampled, y_pred_train_proba)
        train_scores.append(train_auc)

        # Evaluasi pada data uji
        y_pred_proba = model.predict_proba(X_test)[:, 1] if model_name != 'SVM' else model.decision_function(X_test)
        test_auc = roc_auc_score(y_test, y_pred_proba)
        test_scores.append(test_auc)

        print(f'Param value: {param_value}; Train AUC: {train_auc}; Test AUC: {test_auc}')

    plt.plot(param_values, train_scores, label='Train')
    plt.plot(param_values, test_scores, label='Test')
    plt.xlabel(param_name)
    plt.ylabel('AUC')
    plt.title(f'Learning Curve - Hyperparameter {param_name} - {model_name}')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import RandomizedSearchCV

# Definisikan parameter grid untuk Randomized Search dengan penyesuaian
param_dist = {
    'n_estimators': [50, 100, 150],  # Menurunkan jumlah estimator
    'max_depth': [3, 4],  # Menurunkan max_depth
    'learning_rate': [0.01, 0.05, 0.1],  # Menurunkan learning_rate untuk pelatihan yang lebih stabil
    'subsample': [0.7, 0.8],  # Menurunkan subsample
    'colsample_bytree': [0.7, 0.8],  # Menurunkan colsample_bytree
    'min_child_weight': [2, 3],  # Meningkatkan min_child_weight
    'gamma': [0.1, 0.2],  # Meningkatkan gamma
    'scale_pos_weight': [3, 4],  # Meningkatkan scale_pos_weight untuk meningkatkan recall
    'reg_alpha': [0.1, 1.0, 5.0],  # Meningkatkan reg_alpha
    'reg_lambda': [0.1, 1.0, 5.0]  # Meningkatkan reg_lambda
}

# Inisiasi model XGBoost
xgb = XGBClassifier(random_state=42)

# Inisiasi Randomized Search dengan cross-validation
random_search = RandomizedSearchCV(
    estimator=xgb,
    param_distributions=param_dist,
    n_iter=30,  # Kurangi jumlah iterasi untuk mempercepat pencarian
    scoring='roc_auc',
    cv=5,
    n_jobs=-1,
    verbose=1,
    random_state=42  # Set random_state untuk hasil yang konsisten
)

# Latih Randomized Search dengan data pelatihan
random_search.fit(X_train_resampled, y_train_resampled)

# Dapatkan parameter terbaik
best_params = random_search.best_params_
print(f'Best parameters found: {best_params}')

# Evaluasi model dengan parameter terbaik
best_xgb3 = random_search.best_estimator_

# Evaluasi model
eval_classification(best_xgb3, X_train_resampled, y_train_resampled, X_test, y_test)
```
